問題105

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        break
        print('.')
```

問題106

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        continue
        print('.')
```

↑continueがあると強制的に最初に戻る。それによりドットが表示されなくなりループし続ける
